```python
# ----------------- Handle missing values appropriately (e.g., imputation, deletion) --------------------

df.isnull().sum()
```

```python
# Option 1: Imputation with 0

df['LB']=df['LB'].fillna(0)
df['LB'].isnull().sum()

df['AC']=df['AC'].fillna(0)
df['AC'].isnull().sum()

df['DS']=df['DS'].fillna(0)
df['DS'].isnull().sum()

df['DP']=df['DP'].fillna(0)
df['DP'].isnull().sum()

df['ALTV']=df['ALTV'].fillna(0)
df['ALTV'].isnull().sum()

df['MLTV']=df['MLTV'].fillna(0)
df['MLTV'].isnull().sum()

df['Width']=df['Width'].fillna(0)
df['Width'].isnull().sum()

df['Tendency']=df['Tendency'].fillna(0)
df['Tendency'].isnull().sum()

df['NSP']=df['NSP'].fillna(0)
df['NSP'].isnull().sum()


df.isnull().sum()
```

### LB Detect and treat outliers if necessary

```python
plt.figure(figsize=(5,5))
plt.boxplot(df['LB'])
plt.title("LB - Boxplot before outlier treatment")
plt.show()
```

```python
# ---- Detecting outliers ---- Using IQR ------------------------------------------

Q1=df['LB'].quantile(0.25)
Q3=df['LB'].quantile(0.75)

IQR=Q3-Q1

le=Q1-1.5*IQR
ue=Q3+1.5*IQR

print("IQR:",IQR)
print("Lower End:",le)
print("Upper End:",ue)
```

### AC Detect and treat outliers if necessary

```python
plt.figure(figsize=(5,5))
plt.boxplot(df['AC'])
plt.title("AC - Boxplot before outlier treatment")
plt.show()
```

```python
# ---- Detecting outliers ---- Using IQR ------------------------------------------

Q1=df['AC'].quantile(0.25)
Q3=df['AC'].quantile(0.75)

IQR=Q3-Q1

le=Q1-1.5*IQR
ue=Q3+1.5*IQR

print("IQR:",IQR)
print("Lower End:",le)
print("Upper End:",ue)
```

```python
# ---- Treating outliers ------- with Upper End (ue) and Lower End (le) ------------------------------------------

df.AC[df['AC']>ue]=ue
df.AC[df['AC']<le]=le
```
